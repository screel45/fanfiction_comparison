This function will turn the BookNLP html files into txt files that are ready to be imported directly into excel.

How to use:

For Book Canons (one file at a time):
1. Unfortunately I couldn't figure out a good way to do all the canons at once because the naming wouldn't be right for the output files for the different canons/fandoms. So we have to do each canon separately, but it's really fast so doing 6 shouldn't be too terrible.

2. Make sure you replace canon_or_not with True in the function arguments.

3. The first argument is a folder or file path, for the canon you just need the canon's individual booknlp html file path.

4. You can put whatever fandom name you want to be in the filename, it's just for your identification of the output file.

5. It will create the output in whatever directory you're currently sitting in. It names the file as whatever fandom name you put as an argument + _Canon_Characters.txt. You can edit the code if you'd like to have different file names!

For Fanfiction Stories (iterating through a whole folder):
1. Make sure all your FF stories are in separate folders by fandom, e.g. folder 'HP_FF' has files FF_01, FF_02..FF_25.

2. For the folder_or_file_path argument, put in the full path to the folder w/ your stories inside.

3. Replace canon_or_not with False.

4. Create a new folder where you want the outputs to go, and put the full path for the argument output_folder_for_FF

4. It will iterate through each file in the input folder, extract only the total character count and the first title of the character.

5. It then writes these to a new txt file in which the total count is tab separated from the character name.

6. It might not do the files in the exact order that they're in inside your input folder (apparently this is a property of your computer's File System, whatever that means), but to keep them consistent with the naming that you did for the fanfiction scraper I set the new filename to include the exact filename from each input html FF file plus _Characters.txt at the end to identify it as the final output. 

7. It will write the output files to your specified output folder.

8. You can import the txt files into excel (or sheets I'm guessing) making sure that you indicate that it's a tab-delimited file.


```python
'''This function takes the Book NLP output html file, cleans it, and outputs a txt file that only includes
the total character count and the character's most-mentioned title'''

from __future__ import division, unicode_literals 
import codecs
from bs4 import BeautifulSoup
import re
import os

def BookNLP_2_text(input_folder_or_file_path, fandom, canon_or_not, output_folder_for_FF):
    if canon_or_not == True:
        with codecs.open(input_folder_or_file_path,'r', 'latin-1') as f:
            document= BeautifulSoup(f.read()).get_text()
            document=document[159:]
            lines = re.split('\(|\)',document)
            
            character_list = []
            for element in lines:
                if '\t' in element:
                    character_list.append(element.split('\t'))
                    
            with open(str(fandom) + "_Canon_Characters.txt", "w") as outf:
                for line in character_list:
                    outf.write(line[0])
                    outf.write("\t")
                    outf.write(line[1])
                    outf.write("\n")
                    
    if canon_or_not==False:
        folderpath = input_folder_or_file_path
        filenames = os.listdir(folderpath)
        filenames = [file for file in filenames if not file.startswith('.')]
        for file in filenames:
            with codecs.open(os.path.join(folderpath,file),'r', 'latin-1') as f:
                document= BeautifulSoup(f.read()).get_text()
                document=document[159:]
                lines = re.split('\(|\)',document)
            
                character_list = []
                for element in lines:
                    if '\t' in element:
                        character_list.append(element.split('\t'))

                with open(os.path.join(output_folder_for_FF, str(fandom) + '_FF' + str(file) + "_Characters.txt"), "w") as outf:
                    for line in character_list:
                        outf.write(line[0])
                        outf.write("\t")
                        outf.write(line[1])
                        outf.write("\n")
```
